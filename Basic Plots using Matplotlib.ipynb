```python
import matplotlib.pyplot as plt
```

```python
values = [10,5,15,18,25]
plt.hist(values)
plt.title('Basic hist plot')
plt.xlabel('x')
plt.ylabel('y')
```

```python
import matplotlib.pyplot as plt
values = [10,5,15,18,25]
plt.hist(values,color='r',width=2.5)
plt.title('Basic hist plot')
plt.xlabel('x')
plt.ylabel('y')
```

```python
import matplotlib.pyplot as plt
values = [10,5,15,18,25]
plt.bar(values,[0.5,1,1.5,2,2.5],color='g',width=2)
plt.title('Basic bar plot')
plt.xlabel('x')
plt.ylabel('y')
```

```python
import matplotlib.pyplot as plt
values = [10,5,15,18,25]
plt.pie(values,labels=['a','b','c','d','e'],explode=[0,0,0.2,0,0])
plt.title('Basic pie chart')
```
